```python
query = {'work_year' : 2023} #a query that recalls all data in this cleaned dataset.

# Use count_documents to display the number of documents in the result
result = datascijobs.count_documents(query)
# Display the first document in the results using pprint
pprint(result)
```

```python
cleaned_df = pd.DataFrame(datascijobs.find(query))
cleaned_df.head(15)
```

```python
import statsmodels.formula.api as sm #statsmodels package to create a model using the dependent and independent variables 
model =  sm.ols(' salary ~  company_size + experience_level ', data=cleaned_df).fit() #look at company size and enperience level separately in a linear model 
model.summary() #give output of p-values and the coefficients estimating salaries 
```

```python
model =  sm.ols(' salary ~  company_size*experience_level ', data=cleaned_df).fit() #look at interaction between company size and enperience level (instead of a + between variables, it's a *)
model.summary2() #the p>/t/ column for p-values 
```

```python
import statsmodels.formula.api as sm #keep using statsmodels
job_cat_model =  sm.ols(' salary ~  job_category ', data=cleaned_df).fit() #use a singular independent variable to see how many of the categories are statistically significant
job_cat_model.summary() #summary shows all categories but Cloud & Database met the p <0.05 level of significance 
```
